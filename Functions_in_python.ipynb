```python
def multiplier(factor):
    def multiply(x):
        return x * factor
    return multiply

double = multiplier(2)
triple = multiplier(3)

print(double(5))
print(triple(5))
print(double(10))
```

```python
def divide(a, b):
    return a / b

# This crashes:
# print(divide(10, 0))
print(divide(10, "x"))
```

```python
data = [1, 2, 3, 4, 5]

result = list(map(lambda x: x ** 2, data))
print(result)

doubled = list(map(lambda x: x * 2, data))
print(doubled)
```
